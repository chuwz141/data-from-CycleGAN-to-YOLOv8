```python
import os
import numpy as np
import cv2
import torch
import lpips
import torchvision.models as models
import torchvision.transforms as transforms
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
from scipy.linalg import sqrtm
from PIL import Image
from tqdm import tqdm

# **THƯ MỤC ẢNH**
real_images_folder = "/content/drive/MyDrive/G01/Data/Vietnamese_vehicle.v3-2023-02-01-5-31pm.yolov8/test/images"
fake_images_folder = "/content/drive/MyDrive/pytorch-CycleGAN-and-pix2pix/results/day2night_cyclegan/test_latest/nighttime_transformed"

# **LOAD MÔ HÌNH FID**
inception_model = models.inception_v3(pretrained=True, transform_input=False)
inception_model.fc = torch.nn.Identity()  # Bỏ layer cuối cùng
inception_model.eval()

# **LOAD MÔ HÌNH LPIPS**
loss_fn = lpips.LPIPS(net='vgg')

# **CHUẨN BỊ BỘ CHUYỂN ĐỔI ẢNH**
transform_fid = transforms.Compose([
    transforms.Resize((299, 299)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

transform_lpips = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

# **LẤY DANH SÁCH ẢNH (KHÔNG XÉT PHẦN MỞ RỘNG)**
real_images_dict = {os.path.splitext(f)[0]: f for f in os.listdir(real_images_folder)}
fake_images_dict = {os.path.splitext(f)[0]: f for f in os.listdir(fake_images_folder)}

# **LƯU KẾT QUẢ**
psnr_scores, ssim_scores, lpips_scores, real_features, fake_features = [], [], [], [], []
real_features, fake_features = [], []

# **DUYỆT ẢNH VÀ TÍNH METRICS**
for base_name, fake_filename in tqdm(fake_images_dict.items(), desc="Processing images"):
    if base_name in real_images_dict:  # Nếu có ảnh gốc tương ứng
        real_filename = real_images_dict[base_name]

        real_path = os.path.join(real_images_folder, real_filename)
        fake_path = os.path.join(fake_images_folder, fake_filename)

        # **Đọc ảnh với OpenCV để tính PSNR & SSIM**
        real_img_cv = cv2.imread(real_path)
        fake_img_cv = cv2.imread(fake_path)

        # Kiểm tra kích thước ảnh có giống nhau không
        if real_img_cv.shape != fake_img_cv.shape:
            print(f"⚠️ Cảnh báo: Ảnh {real_filename} và {fake_filename} có kích thước khác nhau!")
            continue

        # Chuyển ảnh sang grayscale
        real_gray = cv2.cvtColor(real_img_cv, cv2.COLOR_BGR2GRAY)
        fake_gray = cv2.cvtColor(fake_img_cv, cv2.COLOR_BGR2GRAY)

        # **Tính PSNR**
        psnr_value = psnr(real_img_cv, fake_img_cv, data_range=255)
        psnr_scores.append(psnr_value)

        # **Tính SSIM**
        ssim_value = ssim(real_gray, fake_gray, data_range=255)
        ssim_scores.append(ssim_value)

        # **Tính LPIPS**
        real_img_tensor = transform_lpips(Image.open(real_path).convert("RGB")).unsqueeze(0)
        fake_img_tensor = transform_lpips(Image.open(fake_path).convert("RGB")).unsqueeze(0)
        lpips_value = loss_fn(real_img_tensor, fake_img_tensor).item()
        lpips_scores.append(lpips_value)

        # **Trích xuất đặc trưng FID**
        real_img_fid = transform_fid(Image.open(real_path).convert("RGB")).unsqueeze(0)
        fake_img_fid = transform_fid(Image.open(fake_path).convert("RGB")).unsqueeze(0)

        with torch.no_grad():
            real_features.append(inception_model(real_img_fid).cpu().numpy().flatten())
            fake_features.append(inception_model(fake_img_fid).cpu().numpy().flatten())

# **TÍNH TOÁN FID**
real_features = np.array(real_features)
fake_features = np.array(fake_features)

mu_real, sigma_real = real_features.mean(axis=0), np.cov(real_features, rowvar=False)
mu_fake, sigma_fake = fake_features.mean(axis=0), np.cov(fake_features, rowvar=False)

fid_value = np.sum((mu_real - mu_fake) ** 2) + np.trace(sigma_real + sigma_fake - 2 * sqrtm(sigma_real @ sigma_fake).real)

# **IN KẾT QUẢ**
print("\n========== 📊 KẾT QUẢ ĐÁNH GIÁ 📊 ==========")
print(f"✅ Trung bình PSNR  : {np.mean(psnr_scores):.2f}")
print(f"✅ Trung bình SSIM  : {np.mean(ssim_scores):.4f}")
print(f"✅ Trung bình LPIPS : {np.mean(lpips_scores):.4f}")
print(f"✅ FID Score       : {fid_value:.2f}")
```
