5. Kiá»ƒm tra mÃ´ hÃ¬nh sau huáº¥n luyá»‡n

```python
import cv2
from ultralytics import YOLO
import os
from tqdm import tqdm

# Giáº£ sá»­ model Ä‘Ã£ Ä‘Æ°á»£c load, náº¿u chÆ°a thÃ¬ thÃªm dÃ²ng nÃ y
# model = YOLO('yolov8n.pt')  # Thay báº±ng model cá»§a báº¡n

# ÄÆ°á»ng dáº«n thÆ° má»¥c Ä‘áº§u vÃ o vÃ  Ä‘áº§u ra
input_folder = "/content/drive/MyDrive/Vietnamese_vehicle.v3-2023-02-01-5-31pm.yolov8/test/images/"  # ThÆ° má»¥c chá»©a áº£nh vÃ  video
output_folder = "/content/drive/MyDrive/Vietnamese_vehicle.v3-2023-02-01-5-31pm.yolov8/results"
os.makedirs(output_folder, exist_ok=True)

# Äá»‹nh dáº¡ng file Ä‘Æ°á»£c há»— trá»£
image_extensions = ('.jpg', '.jpeg', '.png')
video_extensions = ('.mp4', '.avi', '.mov')

# Láº¥y danh sÃ¡ch táº¥t cáº£ file trong thÆ° má»¥c
all_files = [f for f in os.listdir(input_folder) if f.lower().endswith(image_extensions + video_extensions)]

# Xá»­ lÃ½ tá»«ng file
for file_name in tqdm(all_files, desc="Processing files"):
    try:
        file_path = os.path.join(input_folder, file_name)
        output_path = os.path.join(output_folder, f"result_{file_name}")

        # Kiá»ƒm tra xem file lÃ  áº£nh hay video
        if file_name.lower().endswith(image_extensions):
            # Xá»­ lÃ½ áº£nh
            results = model(file_path)
            # LÆ°u áº£nh káº¿t quáº£
            results[0].save(filename=output_path)
            print(f"ÄÃ£ xá»­ lÃ½ áº£nh: {file_name}")

        elif file_name.lower().endswith(video_extensions):
            # Xá»­ lÃ½ video
            cap = cv2.VideoCapture(file_path)

            # Láº¥y thÃ´ng tin video
            fps = int(cap.get(cv2.CAP_PROP_FPS))
            width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
            height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

            # Táº¡o video writer
            fourcc = cv2.VideoWriter_fourcc(*'mp4v')
            out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

            # Xá»­ lÃ½ tá»«ng frame
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                # Dá»± Ä‘oÃ¡n vÃ  váº½ káº¿t quáº£ lÃªn frame
                results = model(frame)
                annotated_frame = results[0].plot()
                out.write(annotated_frame)

            # Giáº£i phÃ³ng tÃ i nguyÃªn cho video
            cap.release()
            out.release()
            print(f"ÄÃ£ xá»­ lÃ½ video: {file_name}")

    except Exception as e:
        print(f"Lá»—i khi xá»­ lÃ½ {file_name}: {str(e)}")

cv2.destroyAllWindows()
print("HoÃ n táº¥t xá»­ lÃ½ táº¥t cáº£ file!")
```

```python
from ultralytics import YOLO

# Load mÃ´ hÃ¬nh YOLOv8 Ä‘Ã£ train
model = YOLO("/content/drive/MyDrive/Vietnamese_vehicle.v3-2023-02-01-5-31pm.yolov8/yolov8n.pt")  # Thay báº±ng Ä‘Æ°á»ng dáº«n tá»›i file model cá»§a báº¡n

# Cháº¡y validation trÃªn táº­p dá»¯ liá»‡u
results = model.val(data="/content/drive/.shortcut-targets-by-id/1CeYh-ulsyyWG3e_lYO4Wb0Og1ThxLwZL/Vietnamese_vehicle.v3-2023-02-01-5-31pm.yolov8/data.yaml", imgsz=640)

# TrÃ­ch xuáº¥t vÃ  hiá»ƒn thá»‹ metrics trung bÃ¬nh
print("=== Validation Metrics ===")
print(f"Precision (P): {results.box.p.mean():.4f}")  # Trung bÃ¬nh Precision
print(f"Recall (R): {results.box.r.mean():.4f}")   # Trung bÃ¬nh Recall
print(f"mAP@0.5: {results.box.map50:.4f}")        # mAP@0.5 trung bÃ¬nh
print(f"mAP@0.5:0.95: {results.box.map:.4f}")     # mAP@0.5:0.95 trung bÃ¬nh

# Hiá»ƒn thá»‹ metrics cho tá»«ng class
print("\n=== Per-Class Metrics ===")
for i, class_name in results.names.items():
    if i < len(results.box.maps):  # Äáº£m báº£o khÃ´ng vÆ°á»£t quÃ¡ sá»‘ class cÃ³ metrics
        print(f"Class: {class_name}")
        print(f"  Precision: {results.box.p[i]:.4f}")
        print(f"  Recall: {results.box.r[i]:.4f}")
        print(f"  mAP@0.5: {results.box.ap50[i]:.4f}")  # DÃ¹ng ap50 thay vÃ¬ map50
        print(f"  mAP@0.5:0.95: {results.box.maps[i]:.4f}")

# Hiá»ƒn thá»‹ tá»‘c Ä‘á»™ xá»­ lÃ½
print("\n=== Speed Metrics ===")
print(f"Preprocess: {results.speed['preprocess']:.2f} ms")
print(f"Inference: {results.speed['inference']:.2f} ms")
print(f"Postprocess: {results.speed['postprocess']:.2f} ms")
```
