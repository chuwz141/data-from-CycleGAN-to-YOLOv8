5. Kiểm tra mô hình sau huấn luyện

```python
import cv2
from ultralytics import YOLO
import os
from tqdm import tqdm

# Giả sử model đã được load, nếu chưa thì thêm dòng này
# model = YOLO('yolov8n.pt')  # Thay bằng model của bạn

# Đường dẫn thư mục đầu vào và đầu ra
input_folder = "/content/drive/MyDrive/Vietnamese_vehicle.v3-2023-02-01-5-31pm.yolov8/test/images/"  # Thư mục chứa ảnh và video
output_folder = "/content/drive/MyDrive/Vietnamese_vehicle.v3-2023-02-01-5-31pm.yolov8/results"
os.makedirs(output_folder, exist_ok=True)

# Định dạng file được hỗ trợ
image_extensions = ('.jpg', '.jpeg', '.png')
video_extensions = ('.mp4', '.avi', '.mov')

# Lấy danh sách tất cả file trong thư mục
all_files = [f for f in os.listdir(input_folder) if f.lower().endswith(image_extensions + video_extensions)]

# Xử lý từng file
for file_name in tqdm(all_files, desc="Processing files"):
    try:
        file_path = os.path.join(input_folder, file_name)
        output_path = os.path.join(output_folder, f"result_{file_name}")

        # Kiểm tra xem file là ảnh hay video
        if file_name.lower().endswith(image_extensions):
            # Xử lý ảnh
            results = model(file_path)
            # Lưu ảnh kết quả
            results[0].save(filename=output_path)
            print(f"Đã xử lý ảnh: {file_name}")

        elif file_name.lower().endswith(video_extensions):
            # Xử lý video
            cap = cv2.VideoCapture(file_path)

            # Lấy thông tin video
            fps = int(cap.get(cv2.CAP_PROP_FPS))
            width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
            height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

            # Tạo video writer
            fourcc = cv2.VideoWriter_fourcc(*'mp4v')
            out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

            # Xử lý từng frame
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                # Dự đoán và vẽ kết quả lên frame
                results = model(frame)
                annotated_frame = results[0].plot()
                out.write(annotated_frame)

            # Giải phóng tài nguyên cho video
            cap.release()
            out.release()
            print(f"Đã xử lý video: {file_name}")

    except Exception as e:
        print(f"Lỗi khi xử lý {file_name}: {str(e)}")

cv2.destroyAllWindows()
print("Hoàn tất xử lý tất cả file!")
```

```python
from ultralytics import YOLO

# Load mô hình YOLOv8 đã train
model = YOLO("/content/drive/MyDrive/Vietnamese_vehicle.v3-2023-02-01-5-31pm.yolov8/yolov8n.pt")  # Thay bằng đường dẫn tới file model của bạn

# Chạy validation trên tập dữ liệu
results = model.val(data="/content/drive/.shortcut-targets-by-id/1CeYh-ulsyyWG3e_lYO4Wb0Og1ThxLwZL/Vietnamese_vehicle.v3-2023-02-01-5-31pm.yolov8/data.yaml", imgsz=640)

# Trích xuất và hiển thị metrics trung bình
print("=== Validation Metrics ===")
print(f"Precision (P): {results.box.p.mean():.4f}")  # Trung bình Precision
print(f"Recall (R): {results.box.r.mean():.4f}")   # Trung bình Recall
print(f"mAP@0.5: {results.box.map50:.4f}")        # mAP@0.5 trung bình
print(f"mAP@0.5:0.95: {results.box.map:.4f}")     # mAP@0.5:0.95 trung bình

# Hiển thị metrics cho từng class
print("\n=== Per-Class Metrics ===")
for i, class_name in results.names.items():
    if i < len(results.box.maps):  # Đảm bảo không vượt quá số class có metrics
        print(f"Class: {class_name}")
        print(f"  Precision: {results.box.p[i]:.4f}")
        print(f"  Recall: {results.box.r[i]:.4f}")
        print(f"  mAP@0.5: {results.box.ap50[i]:.4f}")  # Dùng ap50 thay vì map50
        print(f"  mAP@0.5:0.95: {results.box.maps[i]:.4f}")

# Hiển thị tốc độ xử lý
print("\n=== Speed Metrics ===")
print(f"Preprocess: {results.speed['preprocess']:.2f} ms")
print(f"Inference: {results.speed['inference']:.2f} ms")
print(f"Postprocess: {results.speed['postprocess']:.2f} ms")
```
